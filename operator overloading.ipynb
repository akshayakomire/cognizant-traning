```python
class Student:
    def __init__(self, marks):
         self.marks = marks
    def __add__(self,other):
         return self.marks+other.marks
s1=Student(80)
s2=Student(75)
print(s1+s2)
```

```python
class Emp:
    def __init__(self,salary):
        self.salary=salary
    def __add__(self,othersalary):
        return self.salary+othersalary.salary
s1=Emp(10000)
s2=Emp(2000)
print(s1+s2)
```
